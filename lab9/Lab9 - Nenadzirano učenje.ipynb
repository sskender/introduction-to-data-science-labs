## Kmeans
  
Pseudokod algoritma Kmeans:  
1. Nasumičnim odabirom inicijaliziraj k početnih centara grupa.
2. Svaku instancu podatka x pridijeli u grupu čiji je centar najbliži.
3. Osvježi vrijednosti centara grupa tako da budu srednja vrijednost svih podataka koji su pridijeljeni toj grupi.
4. Ako su se centri promijenili, vrati se na korak 2, inače završi.

Kmeans algoritmu moramo unaprijed zadati broj grupa. Isprobajmo nekoliko vrijednosti.

```python
MIN_CLUSTERS = 2
MAX_CLUSTERS = 10

# Compute number of row and columns
COLUMNS = 3
ROWS = math.ceil((MAX_CLUSTERS-MIN_CLUSTERS)/COLUMNS)
fig, axs = plt.subplots(ROWS, COLUMNS, figsize=(10,8), sharey=True, sharex=True)

# Plot the clusters
for n_clusters in range(MIN_CLUSTERS, MAX_CLUSTERS+1):
    current_column = (n_clusters-MIN_CLUSTERS)%COLUMNS
    current_row = (n_clusters-MIN_CLUSTERS)//COLUMNS
    # Get the axis where to add the plot
    ax = axs[current_row, current_column]
    # Cluster the data with the current number of clusters
    kmean = KMeans(n_clusters=n_clusters, random_state=42).fit(X)
    # Plot the data by using the labels as color
    ax.scatter(X[:,0], X[:,1], c=kmean.labels_, alpha=0.6)
    ax.set_title("%s clusters"%n_clusters)
    ax.set_xlabel("Feature 1")
    ax.set_ylabel("Feature 2")
    # Plot the centroids
    for c in kmean.cluster_centers_:
        ax.scatter(c[0], c[1], marker="+", color="red")

plt.tight_layout()
```

Kako izabrati najbolji k?  
  - Metoda "lakta"
  - Korištenjem metrika nenadziranog učenja - Silhouette score

#### Metoda "lakta"

```python
def plot_sse(features_X, start=2, end=11):
    sse = []
    for k in range(start, end):
        # Assign the labels to the clusters
        kmeans = KMeans(n_clusters=k, random_state=10).fit(features_X)
        sse.append({"k": k, "sse": kmeans.inertia_})

    sse = pd.DataFrame(sse)
    # Plot the data
    plt.plot(sse.k, sse.sse)
    plt.xlabel("K")
    plt.ylabel("Sum of Squared Errors")
    
plot_sse(X)
```

Tražimo K nakon kojeg je pogreška stabilna i (skoro) konstantna. U ovom slučaju to je K=4.

#### Silhouette score

```python
silhouettes = []

# Try multiple k
for k in range(2, 11):
    # Cluster the data and assigne the labels
    labels = KMeans(n_clusters=k, random_state=10).fit_predict(X)
    # Get the Silhouette score
    score = silhouette_score(X, labels)
    silhouettes.append({"k": k, "score": score})
    
# Convert to dataframe
silhouettes = pd.DataFrame(silhouettes)

# Plot the data
plt.plot(silhouettes.k, silhouettes.score)
plt.xlabel("K")
plt.ylabel("Silhouette score")
```

Kod ove metode tražimo maksimalni iznos Silhouette scorea, što nam daje K=3. Pogledom na podatke vidimo da su dvije grupe bliske te zbog toga Silhouette score sugerira da je K=3 bolji izbor.

```python
fig, axs = plt.subplots(1, 1, figsize=(4,4), sharey=True)

# Plot the clusters with K = 4
labels = KMeans(n_clusters=4, random_state=0).fit_predict(X)
axs.scatter(X[:,0], X[:,1], c=labels, alpha=0.6)
```

## DBSCAN  
  
Za razliku od Kmeans algoritma, DBSCAN sam određuje koliko grupa ima u podatcima.
  
Pseudokod algoritma:  
1. Uzmi proizvoljan zapis iz liste neobrađenih zapisa (pri pokretanju algoritma to su svi zapisi).
2. Ako u radijusu eps ima minimalno MinPts drugih zapisa, trenutni zapis označi kao središnji zapis i dodaj u listu zapisa trenutne grupe. Sve zapise unutar eps radijusa od trenutnog zapisa dodaj na listu potencijalnih zapisa trenutne grupe.
3. Ako se zapis nalazi u eps radijusu središnjeg zapisa, ali sam nije središnji zapis (korak dva), označen je kao rubni zapis i dodan na listu zapisa trenutne grupe.
4. Ako ima još zapisa u listi potencijalnih zapisa trenutne grupe, uzmi proizvoljni zapis iz te liste i vrati se na korak dva.
5. Ako zapis nije središnji niti rubni, onda on ne pripada nikojoj grupi. Dodaj zapis u listu zapisa bez grupe.
6. Stvori potrebne liste za novu grupu, uzmi proizvoljan zapis iz liste neobrađenih zapisa i vrati se na korak dva.

```python
# Create a list of eps
eps_list = np.linspace(1, 4, 14)

# Compute number of row and columns
COLUMNS = 7
ROWS = math.ceil(len(eps_list)/COLUMNS)

fig, axs = plt.subplots(ROWS, COLUMNS, figsize=(12, 4), sharey=True, sharex=True)

for i in range(0, len(eps_list)):
    eps = eps_list[i]
    
    current_column = i%COLUMNS
    current_row = i//COLUMNS
    
    ax = axs[current_row, current_column]
    labels = DBSCAN(eps=eps).fit_predict(X)
    ax.scatter(X[:,0], X[:,1], c=labels, alpha=0.6)
    ax.set_title("eps = {:.3f}".format(eps))
    
plt.tight_layout()
```

Vizualno nije lako odrediti najbolji parametar eps. Nekoliko je kandidata, ali svaki sa svojim manama. Naravno, u realnom slučaju podatci će biti višedimenzionalni te ovako dobar grafički prikaz vjerojatno neće biti moguć.

```python
silhouettes = []

# Try multiple eps
for eps in eps_list:
    # Cluster the data and assigne the labels
    labels = DBSCAN(eps=eps).fit_predict(X)
    # Get the Silhouette score
    score = silhouette_score(X, labels)
    silhouettes.append({"eps": eps, "score": score})
    
# Convert to dataframe
silhouettes = pd.DataFrame(silhouettes)

# Plot the data
plt.plot(silhouettes.eps, silhouettes.score)
plt.xlabel("eps")
plt.ylabel("Silhouette score")
```

Niti korištenje Silhouette scorea više nije tako jednostavno. Vrijednost metrike brzo uđe u stabilno stanje u kojem ostaje (uz manje oscilacije).  
Iako nije jednostavno za odrediti ispravne parametre, postoje slučajevi u kojima je DBSCAN u velikoj prednosti u odnosu na Kmeans algoritam. Te prednosti bit će predstavljene u nastavku usporedbom više algoritama na više različitih podataka.

## Usporedba algoritama nenadziranog učenja na različitim sintetičkim podatcima  
Osim Kmeansa i DBSCAN-a, prikazat ćemo još šest algoritama koji su implementirani u biblioteci scikit-learn. Dodatno ćemo pojasniti aglomerativno grupiranje i grupiranje maksimizacijom očekivanja.  
  
Aglomerativno grupiranje je bottom-up pristup grupiranju. Izgrađuje se dendogram koji se kasnije "odreže" na željenoj razini kako bi se dobio željeni broj grupa.  
Pseudokod algoritma je:  
1. Svaki pojedini zapis označi kao zasebnu grupu.
2. Izračunaj minimalnu udaljenost među grupama.
3. Spoji dvije grupe s najmanjom udaljenošću među njima.
4. Ako imamo više od jedne grupe, vrati se na korak 2.
5. Dobiveni dendogram "odreži" na željenoj razini.  
  
Algoritam maksimizacije očekivanja spada u grupiranje na temelju distribucije. Također, spada i u grupu mekog grupiranja, što znači da se za svaki zapis daje vjerojatnost da taj zapis pripada određenoj grupi.  
Pseudokod algoritma je:  
1. Nasumičnim odabirom inicijaliziraj početne parametre koji definiraju Gaussove distribucije.
2. Izračunaj vjerojatnosti pripadnosti zapisa svakoj od distribucija.
3. Prilagodi parametre distribucija tako da bolje pašu uz zapise koji su im dodijeljeni.
4. Ponavljaj drugi i treći korak do konvergencije.

```python
from sklearn import cluster, datasets, mixture
from itertools import cycle, islice

n_samples = 1500
noisy_circles = make_circles(n_samples=n_samples, factor=0.5, noise=0.05)
noisy_moons = make_moons(n_samples=n_samples, noise=0.05)
blobs = make_blobs(n_samples=n_samples, random_state=8)
no_structure = np.random.rand(n_samples, 2), None

# Anisotropicly distributed data
random_state = 170
X, y = make_blobs(n_samples=n_samples, random_state=random_state)
transformation = [[0.6, -0.6], [-0.4, 0.8]]
X_aniso = np.dot(X, transformation)
aniso = (X_aniso, y)

# blobs with varied variances
varied = datasets.make_blobs(
    n_samples=n_samples, cluster_std=[1.0, 2.5, 0.5], random_state=random_state
)

# ============
# Set up cluster parameters
# ============
plt.figure(figsize=(9 * 2 + 3, 13))
plt.subplots_adjust(
    left=0.02, right=0.98, bottom=0.001, top=0.95, wspace=0.05, hspace=0.01
)

plot_num = 1

default_base = {
    "quantile": 0.3,
    "eps": 0.3,
    "damping": 0.9,
    "preference": -200,
    "n_neighbors": 10,
    "n_clusters": 3,
    "min_samples": 20,
    "xi": 0.05,
    "min_cluster_size": 0.1,
}

datasets = [
    (
        noisy_circles,
        {
            "damping": 0.77,
            "preference": -240,
            "quantile": 0.2,
            "n_clusters": 2,
            "min_samples": 20,
            "xi": 0.25,
        },
    ),
    (noisy_moons, {"damping": 0.75, "preference": -220, "n_clusters": 2}),
    (
        varied,
        {
            "eps": 0.18,
            "n_neighbors": 2,
            "min_samples": 5,
            "xi": 0.035,
            "min_cluster_size": 0.2,
        },
    ),
    (
        aniso,
        {
            "eps": 0.15,
            "n_neighbors": 2,
            "min_samples": 20,
            "xi": 0.1,
            "min_cluster_size": 0.2,
        },
    ),
    (blobs, {}),
    (no_structure, {}),
]

for i_dataset, (dataset, algo_params) in enumerate(datasets):
    # update parameters with dataset-specific values
    params = default_base.copy()
    params.update(algo_params)

    X, y = dataset

    # normalize dataset for easier parameter selection
    X = StandardScaler().fit_transform(X)

    # estimate bandwidth for mean shift
    bandwidth = cluster.estimate_bandwidth(X, quantile=params["quantile"])

    # connectivity matrix for structured Ward
    connectivity = kneighbors_graph(
        X, n_neighbors=params["n_neighbors"], include_self=False
    )
    # make connectivity symmetric
    connectivity = 0.5 * (connectivity + connectivity.T)

    # ============
    # Create cluster objects
    # ============
    ms = cluster.MeanShift(bandwidth=bandwidth, bin_seeding=True)
    two_means = cluster.MiniBatchKMeans(n_clusters=params["n_clusters"])
    ward = cluster.AgglomerativeClustering(
        n_clusters=params["n_clusters"], linkage="ward", connectivity=connectivity
    )
    spectral = cluster.SpectralClustering(
        n_clusters=params["n_clusters"],
        eigen_solver="arpack",
        affinity="nearest_neighbors",
    )
    dbscan = cluster.DBSCAN(eps=params["eps"])
    optics = cluster.OPTICS(
        min_samples=params["min_samples"],
        xi=params["xi"],
        min_cluster_size=params["min_cluster_size"],
    )
    affinity_propagation = cluster.AffinityPropagation(
        damping=params["damping"], preference=params["preference"], random_state=0
    )
    average_linkage = cluster.AgglomerativeClustering(
        linkage="average",
        affinity="cityblock",
        n_clusters=params["n_clusters"],
        connectivity=connectivity,
    )
    birch = cluster.Birch(n_clusters=params["n_clusters"])
    gmm = mixture.GaussianMixture(
        n_components=params["n_clusters"], covariance_type="full"
    )

    clustering_algorithms = (
        ("MiniBatch\nKMeans", two_means),
        ("Affinity\nPropagation", affinity_propagation),
        ("MeanShift", ms),
        ("Spectral\nClustering", spectral),
        ("Ward", ward),
        ("Agglomerative\nClustering", average_linkage),
        ("DBSCAN", dbscan),
        ("OPTICS", optics),
        ("BIRCH", birch),
        ("Gaussian\nMixture", gmm),
    )

    for name, algorithm in clustering_algorithms:
        t0 = time.time()

        # catch warnings related to kneighbors_graph
        with warnings.catch_warnings():
            warnings.filterwarnings(
                "ignore",
                message="the number of connected components of the "
                + "connectivity matrix is [0-9]{1,2}"
                + " > 1. Completing it to avoid stopping the tree early.",
                category=UserWarning,
            )
            warnings.filterwarnings(
                "ignore",
                message="Graph is not fully connected, spectral embedding"
                + " may not work as expected.",
                category=UserWarning,
            )
            algorithm.fit(X)

        t1 = time.time()
        if hasattr(algorithm, "labels_"):
            y_pred = algorithm.labels_.astype(int)
        else:
            y_pred = algorithm.predict(X)

        plt.subplot(len(datasets), len(clustering_algorithms), plot_num)
        if i_dataset == 0:
            plt.title(name, size=18)

        colors = np.array(
            list(
                islice(
                    cycle(
                        [
                            "#377eb8",
                            "#ff7f00",
                            "#4daf4a",
                            "#f781bf",
                            "#a65628",
                            "#984ea3",
                            "#999999",
                            "#e41a1c",
                            "#dede00",
                        ]
                    ),
                    int(max(y_pred) + 1),
                )
            )
        )
        # add black color for outliers (if any)
        colors = np.append(colors, ["#000000"])
        plt.scatter(X[:, 0], X[:, 1], s=10, color=colors[y_pred])

        plt.xlim(-2.5, 2.5)
        plt.ylim(-2.5, 2.5)
        plt.xticks(())
        plt.yticks(())
        plt.text(
            0.99,
            0.01,
            ("%.2fs" % (t1 - t0)).lstrip("0"),
            transform=plt.gca().transAxes,
            size=15,
            horizontalalignment="right",
        )
        plot_num += 1

plt.show()
```

## Vizualizacija visokodimenzionalnih podataka
Metode transformacije visokodimenzionalnih podataka u niskodimenzionalne podatke pogodne za vizualizaciju:  
  - PCA
  - t-SNE

Generirajmo podatke s 10 značajki i isprobajmo obje metode transformacije za vizualizaciju u nižim dimenzijama.

```python
total_samples = 100

# This create some artifical clusters with standard dev. = 3
X10d, _ = make_blobs(n_samples=total_samples, 
                           centers=3, 
                           cluster_std=3,
                           n_features=10,
                           random_state=0)

print("The features of the first sample are: %s" % X10d[0])
```

```python
X_reduced_tsne = TSNE(n_components=2, random_state=0).fit_transform(X10d)

print("The features of the first sample are: %s" % X_reduced_tsne[0])
```

```python
X_reduced_pca = PCA(n_components=2).fit(X10d).transform(X10d)

print("The features of the first sample are: %s" % X_reduced_pca[0])
```

```python
fig, axs = plt.subplots(1, 2, figsize=(7,3), sharey=True)

# Cluster the data in 3 groups
labels = KMeans(n_clusters=3, random_state=0).fit_predict(X10d)

# Plot the data reduced in 2d space with t-SNE
axs[0].scatter(X_reduced_tsne[:,0], X_reduced_tsne[:,1], c=labels, alpha=0.6)
axs[0].set_title("t-SNE")

# Plot the data reduced in 2d space with PCA
axs[1].scatter(X_reduced_pca[:,0], X_reduced_pca[:,1], c=labels, alpha=0.6)
axs[1].set_title("PCA")
```
