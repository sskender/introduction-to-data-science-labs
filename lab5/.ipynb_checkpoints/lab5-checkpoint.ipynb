# Linearna regresija: Modeliranje dana provedenih u bolnici

- Koriste캖i linearnu regresiju modelirat 캖emo broj dana provedenih u bolnici iz populacije pacijenata.


- Kako bi to izveli trebaju nam dvije komponente
   1. Jednad쬭a koja opisuje model
   2. Podatci
   

- Jednad쬭e se specificiraju koriste캖i [sintaksu](https://patsy.readthedocs.io/en/latest/formulas.html) patsy programskog paketa:
    1. `~` : Odvaja lijevu i desnu stranu
    2. `+` : Stvara uniju prediktora koji se koriste
    3. `:` : Interakcijski termin
    3. `*` : `a * b` je kra캖i zapis za  `a + b + a:b`, i 캜esto je koristan kada 쬰limo uklju캜iti sve interakcije izme캠u skupa varijabli
    
    
- Intercepti se dodaju automatski
- Kategorijske varijable se dodaju koriste캜i termin C(a).
- Za podatke mo쬰mo koristiti pandas!


### Primjer

- Po캜nimo s primjerom iz na코eg skupa podataka. Zanimaju nas dva prediktora: dijabetes i visoki krvni tlak. Ovo su dva prediktora koja 쬰limo upotrijebiti da uklopimo ishod, broj dana provedenih u bolnici, koriste캖i linearnu regresiju.

- Model koji to posti쬰 je formuliran kao:
        time ~ C(diabetes) + C(high_blood_pressure)
        
- Ovaj model mo쬰mo stvoriti pomo캖u smf.ols()

- OLS ozna캜ava obi캜nu linearnu regresiju najmanjih kvadrata

- Dvije komponente: formula i podaci su izri캜ito navedeni.

- Pojmovi u formuli su stupci u pandas podatkovnom okviru. Lako!

```python
# Declares the model
mod = smf.ols(formula='time ~ C(diabetes) + C(high_blood_pressure)', data=df)
```

```python
# Fits the model (find the optimal coefficients, adding a random seed ensures consistency)
res = mod.fit()
```

```python
# Print thes summary output provided by the library.
res.summary()
```

### Puno korisnih informacija pru쬬 se prema zadanim postavkama.

- Zavisna varijabla: vrijeme (broj dana u bolnici)
- Metoda: vrsta modela koji je nau캜en (OLS)
- Nb opa쬬nja: broj to캜aka podataka (299 pacijenata)
- R2: Udio obja코njene varijance
- Popis prediktora
- Za svaki prediktor: koeficijent, standardna pogre코ka koeficijenata, p-vrijednost, 95% intervali povjerenja. Vidimo da je samo visoki krvni tlak zna캜ajan prediktor (p = 0,001), dok dijabetes nije (0,584).
- Upozorenja ako postoje broj캜ani problemi (nadajmo se da ne!) 

### Sada mo쬰mo intrepretirati nau캜eni model

$time = 139.3851 + 4.9059 * diabetes + (-31.8228) * highBloodPressure$

- Koliko je o캜ekivano vrijeme ako pacjent nema dijabetes niti visoki tlak?
- 맚o ako nema dijabetes a ima visoki tlak?
- 맚o ako ima oboje?

1. Osobe koje nemaju dijabetes niti visoki krvni tlak u bolnici borave u prosjeku 139 dana
2. Ljudi koji imaju dijabetes, ali nemaju krvni tlak ostaju 139 + 4,9 dana ~ 144 dana
3. Ljudi koji nemaju dijabetes, ali imaju krvni tlak ostaju 139 - 31 dan ~ 108 dana
4. Osobe s dijabetesom i krvnim tlakom ostaju 139 + 4,9 - 31,8 ~ 112 dana

# Linearna regresija: Modeliranje dana provedenih u bolnici V2

- Jedan od razloga za코to se ozbiljna stanja mogu povezati s kra캖im vremenom provedenim u bolnici je tre캖i 캜imbenik: smrt 游. Pacijenti koji imaju ozbiljno stanje mogli bi provesti manje vremena u bolnici jer umiru.

- Dobijmo bolji osje캖aj 코to se doga캠a modeliraju캖i vrijeme provedeno u bolnici sa smr캖u kao prediktorom.

- Ovaj put 캖emo dodati interakcijske zna캜ajke. 

```python
# we use a*b to add terms: a, b, a:b, and intercept

mod = smf.ols(formula='time ~ C(high_blood_pressure) * C(DEATH_EVENT,  Treatment(reference=0)) + C(diabetes)',
              data=df)

# C(DEATH_EVENT,  Treatment(reference=0)) implies that we are considering the population that did not die!

res = mod.fit()
res.summary()
```

### Interpretacija

- Ovaj model nam omogu캖uje da vidimo da je smrt povezana s manjim brojem dana provedenih u bolnici.
- Primijetite kako je R2 mnogo ve캖i u usporedbi s prethodnim modelom: obja코njeno je vi코e odstupanja u podacima.
- Oni koji imaju visoki krvni tlak ostaju kra캖e (-26 dana u prosjeku), oni koji imaju krvni tlak __i__ umru u prosjeku provode 2,7 dana vi코e u bolnici, iako to nije statisti캜ki zna캜ajno.

# Standardizacija

```python
formula = 'time ~ age + C(high_blood_pressure)'
mod = smf.ols(formula=formula, data=df)
res = mod.fit()
res.summary()
```

맚o mo쬰mo zaklju캜iti iz intercepta? 

> da pacijenti koji nemaju visok pritisak krvi i imaju nula godina ostaju u bolnici ~223 dana

```python
# how we standardize the countinuous variables
columns_to_standardize = [
    "age",
    "creatinine_phosphokinase",
    "ejection_fraction",
    "platelets",
    "serum_creatinine",
    "serum_sodium"
]

for col in columns_to_standardize:
    df[col] = (df[col] - df[col].mean()) / df[col].std()  # standardize column
```

```python
formula = 'time ~ age + C(high_blood_pressure)'
mod = smf.ols(formula=formula, data=df)
res = mod.fit()
res.summary()
```

맚o sada mo쬰mo zaklju캜iti iz intercepta? 

> da pacijenti koji nemaju visok pritisak krvi i imaju prosje캜an broj godina ostaju u bolnici ~140 dana

# Logaritamska transformacija

```python
df = pd.read_csv('data/heart_failure_clinical_records_dataset.csv')

df["logtime"] = np.log(df["time"])
mod = smf.ols(formula='logtime ~ C(high_blood_pressure) * C(DEATH_EVENT,  Treatment(reference=0)) + C(diabetes)',
              data=df)
res = mod.fit()
res.summary()
```

Kako sada mo쬰mo interpretirati ovaj rezultat?

U ovom slu캜aju mo쬰mo re캖i da svaki jedini캜ni porast neke varijable $X_n$ mno쬴 vrijednost ovisne varijable s eksponentom koeficjenta te varijable $\beta_n$.  

Uzmimo koeficjent za dijabetes na primjer (0.0715). 

> $e^{0.0715} = 1.074118$ 

Iz ovog mo쬰mo re캖i da ako sve ostalo ostane isto a vrijednost dijabetesa poraste za jednu jedini캜nu vrijednost tada 캖e pacijent ostati u bolnici 1.07 puta dulje, tj. za 7% dulje.
