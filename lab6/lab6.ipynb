
**Izlaz funkcije `pandas.DataFrame.describe` su vrijednosti deskriptivne statistike. Koristi se i za numeričke i za kategoričke podatake.**

1. *count* - predstavlja ukupni broj vrijednosti u stupcima.
2. *unique* - broji jedinstvene vrijednosti. 
3. *top* - dominantna vrijednost (mod) **D**.
4. *freq* - frekvencija dominantne vrijednosti.
5. *mean* - aritmetička sredina **M**.
6. *std* - standardna devijacija.
7. *min* - najmanja vrijednost.
8. *25%* - **Q<sub>1</sub>**
9. *50%* - **Q<sub>2</sub>** = Centralna vrijednost (medijan) **C**
10. *75%* - **Q<sub>3</sub>**
11. *max* - **Q<sub>4</sub>**, gornja granica.

## z-test

**Nulta hipoteza:** Socio-ekonomska klasa ljudi nije utjecala na stopu preživljavanja.

**Alternativna hipoteza:** Socio-ekonomska klasa ljudi utjecala je na njihovu stopu preživljavanja.

```python
from statsmodels.stats.weightstats import ztest as ztest
```

```python
FirstClass = df.query('Pclass == 1')
ThirdClass = df.query('Pclass == 3')

FirstClass_sample = FirstClass.sample(n=100, random_state=1)
ThirdClass_sample = ThirdClass.sample(n=100, random_state=1)
```

# Naslov
## Naslov

```python
df.groupby('Pclass')['Survived'].mean()
```

```python
FirstClass_sample['Survived'].hist()
```

```python
Survived_normal = np.array([np.mean(df[df["Pclass"]==1].sample(20)["Survived"].values) for i in range(100)])
FirstClass_sample['Survived_normal'] = Survived_normal

Survived_normal_t = np.array([np.mean(df[df["Pclass"]==3].sample(20)["Survived"].values) for i in range(100)])
ThirdClass_sample['Survived_normal'] = Survived_normal_t
```

```python
FirstClass_sample['Survived_normal'].hist()
```

```python
ztest(FirstClass_sample['Survived_normal'], ThirdClass_sample['Survived'], value=0)
```

```python
ztest(FirstClass_sample['Survived_normal'], value=0)
```

```python
ztest(ThirdClass_sample['Survived_normal'], value=0)
```
